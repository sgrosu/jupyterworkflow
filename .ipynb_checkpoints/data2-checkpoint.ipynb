## Unsupervised Analysis of Days of Week

Treating crossings each day to learn about the relationships between various days.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
plt.style.use("seaborn")

from data import get_fremont_data
```

## Get Data

```python
data = get_fremont_data()
```

```python
pivoted = data.pivot_table('Total',index=data.index.time,columns=data.index.date)
pivoted.plot(legend=False,alpha=0.01)
```

## Principal Component Analysis

```python
#pivoted.T.shape
X = pivoted.fillna(0).T.values
```

```python
X2 = PCA(2,svd_solver='full').fit_transform(X)
```

```python
plt.scatter(X2[:,0],X2[:,1])
```

## Unsupervised Clustering

```python
gmm = GaussianMixture(2)
gmm.fit(X)
labels = gmm.predict(X)
plt.scatter(X2[:,0],X2[:,1],c=labels,cmap='rainbow')
plt.colorbar()
```

```python
fig, ax = plt.subplots(1,2,figsize = (14,6))
pivoted.T[labels==0].T.plot(legend=False,alpha=0.1,ax=ax[0])
pivoted.T[labels==1].T.plot(legend=False,alpha=0.1,ax=ax[1])
ax[0].set_title('Purple Cluster')
ax[1].set_title('Red Cluster')
```

## Comparing with Day of Week

```python
dayofweek = pd.DatetimeIndex(pivoted.columns).dayofweek
plt.scatter(X2[:,0],X2[:,1],c=dayofweek,cmap='rainbow')
plt.colorbar()
```

## Analyzing Outliers

The following point are weekdays with a holyday-like pattern

```python
dates = pd.DatetimeIndex(pivoted.columns)
dates[(labels == 0) & (dayofweek < 5)]
```
